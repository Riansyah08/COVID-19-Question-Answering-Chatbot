```python

import pandas as pd

bert_loss = pd.read_csv('Bert Loss.csv')
bert_loss = bert_loss[[bert_loss.columns[0], bert_loss.columns[4]]]
scibert_loss = pd.read_csv('Scibert Loss.csv')
scibert_loss = scibert_loss[[scibert_loss.columns[0], scibert_loss.columns[4]]]
roberta_loss = pd.read_csv('Roberta Loss.csv')
roberta_loss = roberta_loss[[roberta_loss.columns[0], roberta_loss.columns[4]]]
```

```python
import matplotlib.pyplot as plt

plt.plot(bert_loss[bert_loss.columns[0]], bert_loss[bert_loss.columns[1]])
plt.plot(scibert_loss[scibert_loss.columns[0]], scibert_loss[scibert_loss.columns[1]])
plt.plot(roberta_loss[roberta_loss.columns[0]], roberta_loss[roberta_loss.columns[1]])
plt.title("Bert vs Scibert vs Roberta Loss")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.legend(['Bert', 'Scibert', "Roberta"])
```

```python
import json
with open("metrics_bert.json", "r") as f:
    metrics_bert = json.load(f)
with open("metrics_scibert.json", "r") as f:
    metrics_scibert = json.load(f)
with open ("metrics_roberta.json", "r") as f:
    metrics_roberta = json.load(f)
print(metrics_bert)
print(metrics_scibert)
print(metrics_roberta)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming metrics_bert, metrics_scibert, and metrics_roberta are dictionaries containing the values
labels = ["Start Accuracy", "End Accuracy", "Start F1", "End F1"]
bert_values = [metrics_bert["start_accuracy"], metrics_bert["end_accuracy"], metrics_bert["start_f1"], metrics_bert["end_f1"]]
scibert_values = [metrics_scibert["start_accuracy"], metrics_scibert["end_accuracy"], metrics_scibert["start_f1"], metrics_scibert["end_f1"]]
roberta_values = [metrics_roberta["start_accuracy"], metrics_roberta["end_accuracy"], metrics_roberta["start_f1"], metrics_roberta["end_f1"]]

x = np.arange(len(labels))  # label locations
width = 0.25  # width of the bars

# Plotting the bars side by side
fig, ax = plt.subplots()
bar1 = ax.bar(x - width, bert_values, width, label='BERT')
bar2 = ax.bar(x, scibert_values, width, label='SciBERT')
bar3 = ax.bar(x + width, roberta_values, width, label='RoBERTa')

# Adding labels and title
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('BERT vs SciBERT vs RoBERTa Metrics')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Showing the chart
plt.show()
```
